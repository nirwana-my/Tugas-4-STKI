```python
# Inisialisasi hash untuk menyimpan data kontak
contacts = {}

# Fungsi untuk menambahkan kontak ke daftar kontak
def add_contact(name, phone_number, email):
    contacts[name] = {'Phone': phone_number, 'Email': email}
    print(f"Kontak '{name}' berhasil ditambahkan.")

# Fungsi untuk mencari kontak berdasarkan nama
def search_contact(name):
    if name in contacts:
        contact_info = contacts[name]
        print(f"Nama: {name}")
        print(f"Nomor Telepon: {contact_info['Phone']}")
        print(f"Email: {contact_info['Email']}")
    else:
        print(f"Kontak '{name}' tidak ditemukan.")

# Fungsi untuk menampilkan seluruh daftar kontak
def display_contacts():
    if contacts:
        print("Daftar Kontak:")
        for name, info in contacts.items():
            print(f"Nama: {name}")
            print(f"Nomor Telepon: {info['Phone']}")
            print(f"Email: {info['Email']}")
            print("---------------------")
    else:
        print("Daftar kontak kosong.")

# Contoh penggunaan program
add_contact("John Doe", "123456789", "john@example.com")
add_contact("Jane Smith", "987654321", "jane@example.com")

display_contacts()

print("\nCari kontak:")
search_contact("John Doe")
search_contact("Alice")
```

```python
# Definisi kelas untuk simpul dalam pohon biner pencarian
class TreeNode:
    def __init__(self, name, phone_number, email):
        self.name = name
        self.phone_number = phone_number
        self.email = email
        self.left = None
        self.right = None

# Fungsi untuk menambahkan kontak ke dalam pohon biner pencarian
def insert_contact(root, name, phone_number, email):
    if root is None:
        return TreeNode(name, phone_number, email)
    if name < root.name:
        root.left = insert_contact(root.left, name, phone_number, email)
    else:
        root.right = insert_contact(root.right, name, phone_number, email)
    return root

# Fungsi untuk mencari kontak berdasarkan nama dalam pohon biner pencarian
def search_contact_in_tree(root, name):
    if root is None or root.name == name:
        return root
    if name < root.name:
        return search_contact_in_tree(root.left, name)
    return search_contact_in_tree(root.right, name)

# Fungsi untuk mencetak kontak dalam urutan inorder (berurutan)
def inorder_traversal(root):
    if root:
        inorder_traversal(root.left)
        print(f"Nama: {root.name}")
        print(f"Nomor Telepon: {root.phone_number}")
        print(f"Email: {root.email}")
        print("---------------------")
        inorder_traversal(root.right)

# Contoh penggunaan program
if __name__ == "__main__":
    # Membangun pohon biner pencarian untuk kontak
    root = None
    root = insert_contact(root, "John Doe", "123456789", "john@example.com")
    insert_contact(root, "Jane Smith", "987654321", "jane@example.com")

    # Menampilkan kontak dalam urutan inorder
    print("Daftar Kontak:")
    inorder_traversal(root)

    print("\nCari kontak:")
    result = search_contact_in_tree(root, "John Doe")
    if result:
        print("Kontak ditemukan:")
        print(f"Nama: {result.name}")
        print(f"Nomor Telepon: {result.phone_number}")
        print(f"Email: {result.email}")
    else:
        print("Kontak tidak ditemukan.")
```
